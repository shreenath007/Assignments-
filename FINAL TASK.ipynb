### 1. Reading Data from XML format

```python
import pandas as pd 

import xml.etree.ElementTree as etree

tree = etree.parse("customers.xml")

root = tree.getroot()

columns = ["name", "email", "phone", "company", "address", "city", "postal", "country"]

datatframe = pd.DataFrame(columns = columns)

for node in root: 

    name = node.find("name").text if node is not None else None

    email = node.find("email").text if node is not None else None

    phone = node.find("phone").text if node is not None else None
    
    company = node.find("company").text if node is not None else None
    
    address = node.find("address").text if node is not None else None
    
    city = node.find("city").text if node is not None else None
    
    postal = node.find("postal").text if node is not None else None
    
    country = node.find("country").text if node is not None else None

    datatframe = datatframe.append(pd.Series([name, email, phone, company, address, city, postal, country], index = columns), ignore_index = True)
```

```python
# Inserting Timestamp
from datetime import datetime
timestamp = 1545730073
dt_object = datetime.fromtimestamp(timestamp)
print(dt_object)
dt_samples = [dt_object]*len(datatframe)
datatframe['Timestamp'] = dt_samples
```
